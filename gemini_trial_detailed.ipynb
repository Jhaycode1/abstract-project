```python
from chembl_webresource_client.new_client import new_client
import pandas as pd

# 1. SETUP
activity = new_client.activity

# Put your target here. NO QUOTES when we use it later!
my_target = 'CHEMBL287' 

print(f"Connecting to ChEMBL to pull data strictly for: {my_target}...")
target_id = target.filter(target_chembl_id= my_target).only('target_pref_name','target_chembl_id')
print(f"HERE IS IT {target_id}")

# 2. THE CORRECT FILTER
# Notice: target_chembl_id=my_target (Key=Variable. Do not put quotes around my_target)
res = activity.filter(target_chembl_id=my_target) \
              .filter(standard_type__in=['Ki', 'IC50']) \
              .filter(standard_relation="=") \
              .only(['molecule_chembl_id', 'target_pref_name', 'standard_type', 'standard_value', 'standard_units', 'pchembl_value', 'canonical_smiles'])

# Download the first 50 results to test
results = list(res[:50])

if len(results) > 0:
    df = pd.DataFrame(results)
    
    # 3. CLEAN THE DATA
    # Drop rows where pChEMBL couldn't be calculated (the "None" rows)
    df_clean = df.dropna(subset=['pchembl_value'])
    
    print("\n✅ SUCCESS! Here are the different DRUGS (Molecules) hitting your TARGET:")
    # We print just the columns you care about
    print(df_clean[['molecule_chembl_id', 'target_pref_name', 'standard_type', 'standard_value', 'standard_units', 'pchembl_value']].head(10))
    
else:
    print("No results found. Check your internet connection.")
```
