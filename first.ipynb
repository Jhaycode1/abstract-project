```python
from chembl_webresource_client.new_client import new_client
from pandas import pandas as pd 
from chembl_webresource_client.settings import Settings

Settings.Instance().TIMEOUT = 1000


target= new_client.target
activity = new_client.activity
molecule = new_client.molecule
mechanism = new_client.mechanism
target_chembl = 'CHEMBL287'
target_id = target.filter(target_chembl_id= target_chembl)#.only('pref_name','target_type', 'target_prop','target_chembl_id')
#print(f"Locked in as the target {target_id}")
#all activity for the target is here 
res_target = activity.filter(target_chembl_id = target_chembl)\
.filter(standard_type__in = ['Ki','IC50','pchembl_value'])\
.filter(standard_relation = "=")\
.only(['molecule_chembl_id', 'target_pref_name', 'standard_type','molecule_pref_name', 'standard_value', 'standard_units', 'pchembl_value', 'canonical_smiles'])

print(len(res_target))
#print(res)
df_activity = pd.DataFrame.from_records(res_target)
#print(df_activity)
with pd.option_context('display.max_colwidth', None):
    print(f"{df_activity[['molecule_chembl_id','target_pref_name','molecule_pref_name','canonical_smiles', 'type','units', 'value', 'pchembl_value']]}")


#to get the mechanisms of the target and all the cpmpound that bounds as agonist

mec_target= mechanism.filter(target_chembl_id = target_chembl).only('action_type', 'molecule_chembl_id', 'parent_molecule_chembl_id', 'mechanism_comment')
mec_target_df = pd.DataFrame(mec_target)
print(mec_target_df)
```
