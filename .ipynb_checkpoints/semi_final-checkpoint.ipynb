```python
from chembl_webresource_client.new_client import new_client
import pandas as pd 
from chembl_webresource_client.settings import Settings 
Settings.Instance().CACHING = False 
target = new_client.target
mechanism = new_client.mechanism
molecule = new_client.molecule 
activity = new_client.activity 


target_search = target.search('Sigma Non-Opioid Intracellular Receptor 1').only('target_chembl_id', 'target_organism')[:20]
target_df = pd.DataFrame(target_search)
print(target_df)


target_real= 'CHEMBL287'
#finding all molecule thats bids t the target

activities = activity.filter(target_chembl_id= [target_real]
                             , pchembl_value__gte = 7
                             , assay_type = 'B')\
                             .only('pchembl_value', 'molecule_chembl_id', 'parent_molecule_chembl_id', 'molecule_pref_name', 'parent_molecule_pref_name' )
df = pd.DataFrame(activities)
print(df)


df.to_csv('target_file1.csv', index = False )
df.to_txt('target_file1.txt', index = False )
```
