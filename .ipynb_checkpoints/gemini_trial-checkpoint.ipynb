```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

# 1. Initialize the ACTIVITY client (This is where the drugs and numbers live)
activity = new_client.activity
target_id = 'CHEMBL287' # Sigma-1 Receptor

print(f"Connecting to ChEMBL to fetch ALL binding data for {target_id}...")
print("Downloading... This will take a couple of minutes because of pagination. Please wait.")

# 2. The Filter (NO SLICING at the end)
# For a drug repurposing pipeline looking for agonists, EC50 is crucial, so it is included here.
res = activity.filter(target_chembl_id=target_id)\
              .filter(standard_type__in=['Ki', 'IC50', 'EC50'])\
              .filter(standard_relation="=")

# 3. Force the massive download
# Wrapping in list() automatically handles the API pagination behind the scenes
all_data = list(res) 

# 4. Convert to Pandas and Clean
if len(all_data) > 0:
    df_all = pd.DataFrame(all_data)
    
    print(f"\n MASSIVE DOWNLOAD COMPLETE!")
    print(f"Total raw rows downloaded: {len(df_all)}")
    
    # Clean out the rows that don't have actual numbers
    df_clean = df_all.dropna(subset=['standard_value'])
    print(f"Total clean rows with valid numbers: {len(df_clean)}")
    
    # HIGHLY RECOMMENDED: Save this massive dataset to your computer immediately!
    # This way you never have to wait 3 minutes for it to download again.
    df_clean.to_csv("ALL_SIG1R_Compounds.csv", index=False)
    print("\nData saved to 'ALL_SIG1R_Compounds.csv'")
    
    # Show a sneak peek
    cols = ['molecule_chembl_id', 'standard_type', 'standard_value', 'standard_units']
    print("\n--- SNEAK PEEK ---")
    print(df_clean[cols].head())
    
else:
    print("No data found.")
```
