```python
from chembl_webresource_client.new_client import new_client
import pandas as pd 
from chembl_webresource_client.settings import Settings 
Settings.Instance().CACHING = False 
target = new_client.target
mechanism = new_client.mechanism
molecule = new_client.molecule 
activity = new_client.activity 


target_search = target.search('Sigma Non-Opioid Intracellular Receptor 1').only('target_chembl_id', 'target_organism')[:20]
target_df = pd.DataFrame(target_search)
print(target_df)


target_real= 'CHEMBL287'
#finding all that binds to target

activities = activity.filter(target_chembl_id = target_real).only('molecule_chembl_id', 'molecule_pref_name', 'parent_molecule_chembl_id')
activities_df = pd.DataFrame(activities)
print(activities_df)

activities_df.to_csv('activities3.csv', index = False)
print('activities_df saved')


my_molecule= pd.read_csv('activities3.csv', usecols=['parent_molecule_chembl_id'])
molecules = my_molecule['parent_molecule_chembl_id'].tolist()

#print(molecules)



mec = mechanism.filter(molecule_chembl_id__in= molecules).only('action_type', 'molecule_chembl_id', 'parent_molecule_chembl_id')
mec_df = pd.DataFrame(mec)
print(mec_df)
mec_df.to_csv('mechanism3.csv', index = False)
print('mechanism3.csv saved')

mole= pd.read_csv('mechanism3.csv', usecols=['parent_molecule_chembl_id'])
final_mole = mole['parent_molecule_chembl_id'].tolist()

#print(final_mole)

mol= molecule.filter(molecule_chembl_id__in = final_mole).only('molecule_chembl_id', 'max_phase', 'pref_name')
mol_df = pd.DataFrame(mol)
mol_df.to_csv('molecule3.csv', index = False)
print('molecule3.csv saved')

```
